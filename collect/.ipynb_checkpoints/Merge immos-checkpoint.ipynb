```python
# when analyzing search for immo id and check props, maybe same immo / new immo with same id
folder = '/Users/a2/.cache/immodynamics/1666562745' #
import pandas as pd
import os 
import numpy as np
files = os.listdir(folder)
exposes = []
for f in files:
    if f.endswith('.csv'):
        exposes += [f]

table = pd.DataFrame()
temp = -1
for e in exposes:
    try:
        df = pd.read_csv(f'{folder}/{e}')
        for n,i in enumerate(['A+', 'A', 'B', 'C', 'D', 'E', 'F', 'G', 'H']):
            if df.effclass[0]==i:
                temp = n
                break
        else: 
            raise ValueError("NaN")
    except ValueError:
        pass
    else:
        df.drop('effclass', axis=1)
        df['effclass'] = [float(temp)]
        table = pd.concat([table, df], ignore_index=True)
for t in range(len(table['preis'])):
    index = table['preis'][t].find('&')
    if index != -1: 
        temp = table['preis'][t][:index]
        temp = temp.split(",")[0].replace(".","")
    table['preis'][t] = int(temp)    
    table['qm'][t] = int(table['qm'][t])
    table.effclass[t] = float(table.effclass[t])
    




table.to_csv(f"{folder}/main.csv")



print(len(exposes), len(table))
```
